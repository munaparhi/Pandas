There are 4 advanced method to update the value in column---

apply :
Applies  function to each element in a column or row of the DataFrame.
Use: Applies a function to each element in a column or row of the DataFrame.
Works with: DataFrame (column-wise or row-wise) and Series.
When to Use: When you need to apply a function that operates on each row or column independently.

map:
Use: Applies a function to each element in a Series. It also works with dictionaries for value replacement.
Works with: Series only.
When to Use: When transforming values in a single column/Series


applymap:
Use: Applies a function to every element in the entire DataFrame.
Works with: DataFrame only.
When to Use: When every individual element of the DataFrame needs to be updated.

replace:
Use: Replaces specified values in a DataFrame or Series with other values. Can work with a dictionary for flexible replacement.
Works with: DataFrame and Series.
When to Use: When replacing specific values directly without applying functions.
