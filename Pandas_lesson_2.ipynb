We can also do this by simple filtering...  let we want to see the  uses of social media by specific country 

```python
filt=df['Country']=='India'
df.loc[filt,'SocialMedia'].value_counts()
```

```python
df.groupby('Country')['SocialMedia'].value_counts().loc['India']  # if you want to see the results in Percentage then use normalize=True.
```

```python
country_grp['ConvertedComp'].agg(['median','mean'])  #Here we can pass multiple agg fun in list...and if we want to see the
#specific country mean and median just pass .loc['CountryName']
```

```python
filt=df['Country']=='India'
df.loc[filt,'LanguageWorkedWith'].str.contains('Python').sum()
```

```python
#country_grp['LanguageWorkedWith'].str.contains('Python').sum() 
#This is the wrong method
```

The string method only works on series but not in the SeriesGroupBy...So we use apply method so that it will work for every series of the group.

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_py = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_py
```

```python
python_df=pd.concat([country_respondents,country_uses_py],axis=1)
python_df
```

```python
python_df.rename({'Country':'Totalrespondent','LanguageWorkedWith':'pplknowspy'},axis=1,inplace=True)
```

```python
python_df
```

```python
python_df['pct_knows_py'] = (python_df['pplknowspy']/python_df['Totalrespondent'])*100
```

```python
python_df
```

Now we can directly grab the information of a country like below....

```python
python_df.loc['India']
```

# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
df.head()
```

Suppose we want to handle the  custom missing values as NaN value we can do that when we read_csv value by passing the custom missing value in a variable and na_values = variable_name

```python
na_vals=['NA','Missing']

df = pd.read_csv(r"C:\Users\parhi\OneDrive\Desktop\stack-overflow-developer-survey-2019\survey_results_public.csv",na_values=na_vals)
```

```python
df.head()
```

```python
df['YearsCode'].dtypes
```

```python
#df['YearsCode']= df['YearsCode'].astype(float) this will give an error .
```

```python
df['YearsCode'].unique()  # .unique method is used to see all distinct value... Or we can also use value_counts() method but
# here we only need distinct values not the counts of values.
```

```python
df['YearsCode'].replace('Less than 1 year',0,inplace= True )
df['YearsCode'].replace('More than 50 years',51,inplace= True )
```

```python
df['YearsCode']= df['YearsCode'].astype(float)  # if you want to convert it to int value then first handle 'nan' values
```

```python
df['YearsCode'].mean()
```

```python
df['YearsCode'].median()
```
